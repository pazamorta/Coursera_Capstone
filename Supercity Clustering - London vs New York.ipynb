## 1. Obtaining London Tube Stations

We use the open data from TFL and convert the kml file into a geojson https://api-portal.tfl.gov.uk/docs

```python
import kml2geojson
kml2geojson.main.convert('Data/Londonstations.kml', 'Data/', separate_folders=False, style_type=None, style_filename='style.json')
```

Load the geojson 

```python
with open('Data/Londonstations.geojson') as json_data:
    london_data = json.load(json_data)
```

Review the structure of the geojson

```python
stations_data = london_data['features']
stations_data[0]
```

Create a Pandas Dataframe

```python
dataframe_columns = ['Name','City','Latitude','Longitude']
london_dataframe = pd.DataFrame(columns=dataframe_columns)
```

Iterate through the stations data to populate the Dataframe

```python
for data in stations_data:
    station_name = data['properties']['name']
    
    station_latlon = data['geometry']['coordinates']
    station_lat = station_latlon[1]
    station_lon = station_latlon[0]

    
    london_dataframe = london_dataframe.append({
                            'Name': station_name,
                            'City': 'London',
                            'Latitude': station_lat,
                            'Longitude': station_lon}, ignore_index=True)
```

## 4. Using Foursquare API to rasterize the venues in the super-city

Initial values to perform requests on the Foursquare API

```python
CLIENT_ID = '' 
CLIENT_SECRET = '' 
VERSION = '20180605' 
LIMIT = 200
radius = 1000
```

GetNearbyVenues Function to Store Venues in a Neighbourhood

```python
def getNearbyVenues(names, city, latitudes, longitudes, radius=1500):
    
    venues_list=[]
    for name, cty, lat, lng in zip(names, city, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request with a try/except construct to retry if errors occur
        try:
            results = requests.get(url, timeout=5).json()["response"]['groups'][0]['items']
        except requests.exceptions.Timeout:
            try:
                results = requests.get(url, timeout=5).json()["response"]['groups'][0]['items']
            except requests.exceptions.Timeout: 
                pass
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            cty,
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood',
                  'City',                         
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude',
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)

super_city_venues = getNearbyVenues(names=super_city['Name'],
                                   city=super_city['City'],
                                   latitudes=super_city['Latitude'],
                                   longitudes=super_city['Longitude']
                                  )
```

## 5. Super-city Neighborhoods Analysis

On-hot encoding of all venue categories across all neighborhoods. The result is a matrix of the size displayed under the cell below.

```python
# one hot encoding
super_city_onehot = pd.get_dummies(super_city_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
super_city_onehot['Neighborhood'] = super_city_venues['Neighborhood'] 

# set Neighborhood to index
cols_to_order = ['Neighborhood']
new_columns = cols_to_order + (super_city_onehot.columns.drop(cols_to_order).tolist())
super_city_onehot[new_columns]

super_city_onehot.shape
```

We calculate the level of frequency of the different types of venues in each one of the neighborhoods.

```python
super_city_grouped = super_city_onehot.groupby('Neighborhood').mean().reset_index()
super_city_grouped
```

We sort the venues in desceding order of frequency

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Display to 10 venue categories per neighborhood

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = super_city_grouped['Neighborhood']

for ind in np.arange(super_city_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(super_city_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## 6. Neighborhood Clustering

We use K-means to cluster the neighborhoods based on the one-hot encoding matrix with all neighbourhoods and the frequency of each venue type.

First we search for the elbow in the graph applying different values of k

```python
# determine k using elbow method
from sklearn.cluster import KMeans

super_city_grouped_clustering = super_city_grouped.drop('Neighborhood', 1)

Sum_of_squared_distances = []
K = range(1,30)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(super_city_grouped_clustering)
    Sum_of_squared_distances.append(km.inertia_)
    
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

In this case, this is inconclusive since there is no clear change in the descent slope but a relatively smooth convergence. We think k=10 seems to provide a good result and allowing plotting the information.

```python
# set number of clusters
kclusters = 10

super_city_grouped_clustering = super_city_grouped.drop('Neighborhood', 1)

# run k-means clustering
# using k-means++ to initialize the centroids to be distant from each other
# since this can lead to better results than random initialization
kmeans = KMeans(n_clusters=kclusters, init='k-means++', random_state=0).fit(super_city_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

We reconstruct the matrix after applying the clustering label with the information about the neighbourhood.

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

super_city_merged = super_city

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
super_city_merged = super_city_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Name')

super_city_merged.head() # check the last columns!
```
